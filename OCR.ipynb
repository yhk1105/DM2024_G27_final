```python
from ast import main
# api_key = Google ocr key
count = 0
temp1 = list(dt.keys())
for folder in mainls:
  temp = folder.split(".")[0]
  count += 1
  pdf_text = ''
  if folder.endswith('.pdf') and temp in temp2:
    ls = os.listdir("drive/MyDrive/finance_pdfpic/"+folder)
    for i in ls:
      with open("drive/MyDrive/finance_pdfpic/"+folder+"/"+i, "rb") as image_file:
        content = base64.b64encode(image_file.read()).decode("utf-8")
      url = f"https://vision.googleapis.com/v1/images:annotate?key={api_key}"
      headers = {"Content-Type": "application/json"}
      data = {
      "requests": [
        {
          "image": {"content": content},
          "features": [{"type": "TEXT_DETECTION"}]
        }
        ]
      }
      success = False
      retries = 3
      while not success and retries > 0:
        try:
          response = requests.post(url, headers=headers, json=data, timeout=10)
          response.raise_for_status()
          result = response.json()
          text = result['responses'][0].get('textAnnotations', [])
          pdf_text += text[0]['description']
          success = True
        except requests.exceptions.RequestException as e:
          print(f"Error: {e}. Retrying in 5 seconds...")
          retries -= 1
          time.sleep(5)  # 等待 5 秒後重試
        except KeyError:
          print("Unexpected response format. Skipping this page.")
          break
  pdf_text = pdf_text.replace('\n', '')
  name = folder.split(".")[0]
  dt[name] = pdf_text
  print(count)
```
