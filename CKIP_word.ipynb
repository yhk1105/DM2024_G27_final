```python
from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
data_utils.download_data_url("./")
```

```python
ws = WS("./data")
pos = POS("./data")
ner = NER("./data")
```

```python
import json
with open("Correct_Finance_OCR.json", 'rb') as f:
  ins_ref = json.load(f)
from tqdm import tqdm
dt = dict()
for i, v in tqdm(ins_ref.items()):
  temp = i.split(".")
  one = int(temp[0])
  if one >=700:
    a = ws([v])
    b = pos(a)
    c = list(ner(a,b))
    dt[i] = {"ws":a[0], "pos":b[0],"ner":list(c[0])}
with open("ckip_Finance_data.json", "w", encoding="utf-8") as f:
    json.dump(dt, f, ensure_ascii=False, indent=4)
```
